# Employee Name 

```python
data[data['Employee_Name'].isnull()]
```

```python
# we need to change null to Unkown if we want calculate something without go back for the name
data['Employee_Name'].fillna('Unkown',inplace=True)
data
```

```python
data.isnull().sum()
```

```python
Salmean=data.groupby(['Sex']).mean()['Salary']
print(Salmean)
```

```python
# in salary the mean will be replaced with null 

data.loc[(data['Sex']=="F")&(data['Salary'].isnull())]=\
data.loc[(data['Sex']=="F")&(data['Salary'].isnull())].fillna(Salmean['F'])
```

```python

data.loc[(data['Sex']=="M ")&(data['Salary'].isnull())]=\
data.loc[(data['Sex']=="M ")&(data['Salary'].isnull())].fillna(Salmean['M '])
```

```python
data.dropna(subset=['Salary'],inplace=True)
```

```python
data.isnull().sum()
```

```python
data['DOB'].fillna('01-01-2000', inplace=True)
```

```python
data['Sex'].fillna('not_defind', inplace=True)
```

```python
#manger id for Webster Butler is 39
data.loc[data['ManagerID'].isnull(),'ManagerID']=39
```

```python
data.isnull().sum()
```

```python
data['DateofTermination'].fillna('Still Employed', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.loc[data['Absences'].isnull(),'Absences']=0
```
